```python
import pandas as pd

file_path = r'C:\Users\92349\Desktop\Prodigy internship\world bank.csv'

df = pd.read_csv(file_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

selected_year = '2020'

data_to_visualize = df[['Country Name', selected_year]]

data_to_visualize = data_to_visualize.sort_values(by=selected_year, ascending=False)
```

```python


# Create a bar chart
plt.figure(figsize=(15, 8))
plt.bar(data_to_visualize['Country Name'], data_to_visualize[selected_year], color='blue')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title(f'Distribution of Population in {selected_year}')
plt.xticks(rotation=90) 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
# List of specific countries you want to visualize
selected_countries = ['China', 'India', 'United States', 'Indonesia', 'Pakistan']

# Filter the DataFrame for the selected countries and the chosen year
data_to_visualize = df[df['Country Name'].isin(selected_countries)][['Country Name', '2020']]

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
bars = plt.barh(data_to_visualize['Country Name'], data_to_visualize['2020'], color='blue')
plt.xlabel('Population')
plt.ylabel('Country')
plt.title('Population of Selected Countries in 2020')

# Display population values next to each bar
for bar in bars:
    xval = bar.get_width()
    plt.text(xval, bar.get_y() + bar.get_height()/2, f'{int(xval):,}', va='center', ha='left')

plt.show()
```
